````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# 1. Initialize inventory with error handling
def initialize_inventory(products):
    inventory = {}

    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")
                inventory[product] = quantity
                break
            except ValueError as e:
                print(f"Error: {e}")

    return inventory


# 2. Get customer orders with error handling
def get_customer_orders(inventory):
    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders <= 0:
                raise ValueError("Number of orders must be positive.")
            break
        except ValueError as e:
            print(f"Error: {e}")

    customer_orders = []

    for _ in range(num_orders):
        while True:
            try:
                product = input("Enter the name of a product that a customer wants to order: ").strip()

                if product not in inventory:
                    raise ValueError("Product does not exist.")

                if inventory[product] <= 0:
                    raise ValueError("Product is out of stock.")

                customer_orders.append(product)
                break

            except ValueError as e:
                print(f"Error: {e}")

    return customer_orders


# 3. Order statistics (no change needed)
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    unique_products_ordered = len(set(customer_orders))
    percentage_unique = (unique_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_unique


# 4. Update inventory (already correct)
def update_inventory(inventory, customer_orders):
    updated_inventory = inventory.copy()

    for product in customer_orders:
        if product in updated_inventory:
            updated_inventory[product] -= 1

    updated_inventory = {
        product: quantity
        for product, quantity in updated_inventory.items()
        if quantity > 0
    }

    return updated_inventory


# 5. Total price with error handling
def calculate_total_price(customer_orders):
    prices = {}

    for product in set(customer_orders):
        while True:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                prices[product] = price
                break
            except ValueError as e:
                print(f"Error: {e}")

    total_price = sum(prices[product] for product in customer_orders)
    return total_price


# --- RUN PROGRAM ---

inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)

order_statistics = calculate_order_statistics(customer_orders, products)
print("\nOrder Statistics:")
print(f"Total Products Ordered: {order_statistics[0]}")
print(f"Percentage of Unique Products Ordered: {order_statistics[1]}")

updated_inventory = update_inventory(inventory, customer_orders)
print("\nUpdated Inventory:")
for product, quantity in updated_inventory.items():
    print(f"{product}: {quantity}")

total_price = calculate_total_price(customer_orders)
print(f"\nTotal Price: {total_price}")
````
